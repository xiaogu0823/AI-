```python
# from langchain.llms import OpenAI该方法已更新为langchain_openai
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
# from langchain_core.output_parsers import StrPutParser

# base_url="https://api.siliconflow.cn/v1"
# api_key="sk-pnwvnunvbaigqkllfgmvjlashqqhxvsgunrvhjuktowqncdi"
base_url = "https://api.deepseek.com/v1"
api_key = "sk-8615b7b0f4f241239ef4c80a716b7588"
api_base = base_url
api_key = api_key

llm = ChatOpenAI(
    model="deepseek-chat",
    temperature=0,
    openai_api_base = api_base,
    openai_api_key = api_key
)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "你是一个写七言律诗的诗人"),
        ("user", "{input}"),
    ]
)

# 格式化
# output_parser = StrPutParser()
# 通过langchain的链式调用，生成chain
chain = prompt | llm 
# llm.predict('给我找一首诗，叫春晓')
result = chain.invoke(input='给我找一首诗，关于桃花')
print(result)
```
